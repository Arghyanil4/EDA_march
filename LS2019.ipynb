```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
LS_2019_df=pd.read_csv(r"C:\Users\dasar\Documents\Data Science\Naresh IT\Data sets\LS_2.0.csv")
LS_2019_df
```

```python
total_votes = LS_2019_df.groupby('CONSTITUENCY')['TOTAL\nVOTES'].sum()
total_electors = LS_2019_df.groupby('CONSTITUENCY')['TOTAL ELECTORS'].unique().apply(lambda x: x[0])
voter_turnout = round((total_votes / total_electors) * 100,2)
print(voter_turnout)
```

```python
print("count",len(voter_turnout))
print("Min Value",voter_turnout.idxmin(),round(voter_turnout.min(), 2))
print("Max Value",voter_turnout.idxmax(),round(voter_turnout.max(), 2))
print("Mean Value",round(voter_turnout.mean(), 2))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(voter_turnout, bins=200, edgecolor='black')
plt.title('Histogram of Voter Turnout Percentages')
plt.xlabel('Voter Turnout Percentage')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
