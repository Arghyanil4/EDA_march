**Data transformations**
- Generally data has 3 types of distributions
    - Normal distribution
    - Positive Skewed
    - Negative Skewed
- All the maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods, if the data does not follow normality
- We have some transformation methods mentioned below
    - Log Transformation
    - Exponential Transformation
    - Reciprocal Transformation
    - Square root Transformation
    - Power Transformation

**Log Transformation**
- Log transformation means performing logarithm operations on original data
- It is an approach to convert into normal distribution
- Log means natural logarithm base=e


```python
x=2
np.log(x)
```

```python
exp_data = np.random.exponential(size=10000)

# Plot the histograms
plt.figure(figsize=(10, 5))

# Plot for exponential data
plt.subplot(1, 2, 1)
plt.hist(exp_data, bins=20, label='Exponential')
plt.legend()

# Plot for reciprocal data
log_data = np.log(exp_data)
plt.subplot(1, 2, 2)
plt.hist(log_data, bins=20, label='Logarithmic')
plt.legend()

# Show the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

# Generate 10k random exponential numbers
exp_data = np.random.exponential(size=10000)

# Convert single-dimensional array to 2-dimensional list
exp_data_lst = [[k] for k in exp_data]

# Create PowerTransformer object with 'yeo-johnson' method
pt_yj = PowerTransformer(method='yeo-johnson')

# Perform Yeo-Johnson transformation
yeo_data = pt_yj.fit_transform(exp_data_lst)

# Create PowerTransformer object with 'box-cox' method
pt_bc = PowerTransformer(method='box-cox')

# Perform Box-Cox transformation
box_cox_data = pt_bc.fit_transform(exp_data_lst)

# Create subplots with 1 row and 2 columns
fig, axis = plt.subplots(1, 2, figsize=(16, 9))

# Title for Figure
fig.suptitle("Power Transformation Methods")

# Setting titles for subplots
axis[0].set_title("Yeo-Johnson Power Transformation")
sns.histplot(ax=axis[0], data=yeo_data.flatten())

axis[1].set_title("Box-Cox Power Transformation")
sns.histplot(ax=axis[1], data=box_cox_data.flatten())

plt.show()
```
