**Label encoder**
- Label encoder is a package under **scikit learn**
- Scikit - learn is heart of Machine Learning
- In side SKlearn we have method called **Preprocessing**
- Under preprocessing we have **LabelEncoder**
- scikit-learn we will write as sklearn
- any sklearn package we have 3 steps
    - Step-1: Read the package
    - Step-2: Save the package
    - Step-3: Apply the fitransform


```python
from sklearn.preprocessing import LabelEncoder
#Step -2 : Save the package
le = LabelEncoder()
# Step-3: apply fit transform
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
for column in visa_df.select_dtypes(include='object').columns:
    visa_df[column]=le.fit_transform(visa_df[column])
visa_df
```

**np.where**

- np.where applicable only for binary
- np.where will take 3 arguments
   - condition
   - True value
   - False value

```python
visa_df=pd.read_csv(r"C:\Users\dasar\Documents\Data Science\Naresh IT\Data sets\Visadataset.csv")

for column in visa_df.select_dtypes(include='object').columns:
    # Create a mapping dictionary for unique values in the column
    labels = list(visa_df[column].unique())
    label_dict = {label: i for i, label in enumerate(labels)}
    
    # Apply the mapping to the DataFrame using np.where
    for label, idx in label_dict.items():
        visa_df[column] = np.where(visa_df[column] == label, idx, visa_df[column])

visa_df
```
