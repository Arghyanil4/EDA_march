```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the Titanic dataset loaded as 'titanic'
# Get the counts of each passenger class
pClass_counts = titanic['Pclass'].value_counts()

# Create a DataFrame with the counts
pClass_df = pd.DataFrame({'Passenger class': pClass_counts.index, 'Count': pClass_counts.values})

################## Bar-Chart using Matplotlib ########################
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.bar(pClass_df['Passenger class'], pClass_df['Count'])
plt.title("Titanic Passengers")
plt.xlabel("Passenger class")
plt.ylabel("Passenger Count")

################# Bar-Chart using Seaborn ######################
plt.subplot(1, 3, 2)
# Sort the DataFrame by count for the Seaborn plot
pClass_df_sorted = pClass_df.sort_values(by='Count', ascending=False)
sns.barplot(data=pClass_df_sorted, x='Passenger class', y='Count')
plt.title('Counts of Each Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

################# Bar-Chart Value count directly ######################
plt.subplot(1, 3, 3)
# Plotting directly using the value_counts DataFrame
pClass_counts.sort_values(ascending=False).plot(kind='bar')
plt.title("Passenger Class Counts")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
pClass_df_sorted = titanic['Pclass'].value_counts()
keys = pClass_df_sorted.index
values = pClass_df_sorted.values

# Create a pie chart
plt.pie(values, explode=[0.02, 0.02, 0.02], labels=keys, autopct="%0.2f%%", startangle=180, radius=1)
plt.title('Distribution of Passenger Classes on the Titanic')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the Titanic dataset loaded as 'titanic'
# Select only numeric columns
numeric_cols = titanic.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_cols.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, linewidths=0.5)

plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```
